```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from typing import Optional, List
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
class DocumentTextSplitterManager:

    def __init__(self, chunk_size : int = 100,chunk_overlap : int =20,length_function: callable = len):
        self.chunk_size=chunk_size
        self.chunk_overlap=chunk_overlap
        self.length_function=length_function
    
    def get_recursive_splitter(self, separators : Optional[List[str]] = None) -> RecursiveCharacterTextSplitter:
        if separators is None:
            separators = [
                "\n\n",  # Paragraphs
                "\n",    # Lines
                ". ",    # Sentences
                "! ",
                "? ",
                "; ",    # Clauses
                ", ",    # Phrases
                " ",     # Words
                ""       # Characters (last resort)
            ]
        splitter = RecursiveCharacterTextSplitter(
            separators=separators,
            chunk_size=self.chunk_size,
            chunk_overlap=self.chunk_overlap,
            length_function=self.length_function
            )
        logger.info(f"Created recursive character splier"
                    
                    f"size:{self.chunk_size}, overlap: {self.chunk_overlap}")
        return splitter
```

```python
from langchain_classic.embeddings import OpenAIEmbeddings, SentenceTransformerEmbeddings
from typing import Optional
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
class EmbeddingManager:

    def __init__(self, provider: str = 'sentenceTransformer', model_name: Optional[str] = None):
        self.provider = provider
        self.embeddings = self._initialize_embeddings(provider=provider, model_name=model_name) 
        logger.info(f"Initialized {provider} embeddings")

    def _initialize_embeddings(self, provider, model_name: Optional[str]):
        if provider == "openai":
            return self._init_open_ai(model_name)
        elif provider == "sentenceTransformer":
            return self._init_sentence_transformer_embedding(model_name)
        else:
            raise ValueError(f"invalid provider: {provider}. "
                             f"We only support SentenceTransformer")
        
    def _init_open_ai(self, model_name: Optional[str]):  # âœ… Added self
        pass

    def _init_sentence_transformer_embedding(self, model_name: Optional[str]):
        if model_name is None:
            model_name = "thenlper/gte-small"
        return SentenceTransformerEmbeddings(model_name=model_name)
```
